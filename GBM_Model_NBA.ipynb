# Without PCA Implementation by using rolling and normal stats together

```python

import pandas as pd

# Create empty DataFrames for train and test datasets
train_df = pd.DataFrame()
test_df = pd.DataFrame()

# Split each player_id into 80% train and 20% test
for player_id, group in gbm_data.groupby('Player_ID'):  
    # Split into 80% train and 20% test
    split_index = int(len(group) * 0.8)
    train_part = group.iloc[:split_index]  # First 80%
    test_part = group.iloc[split_index:]   # Last 20%
    
    # Append results to the main train and test DataFrames
    train_df = pd.concat([train_df, train_part])
    test_df = pd.concat([test_df, test_part])

# Define features (X) and target (y) for training
X_train = train_df[['PTS', 'MIN_x', 'TOV_x', 'AST_x', 'REB_x', 'FT_PCT', 
                    'PTS_rolling_3', 'AST_rolling_3', 'REB_rolling_3', 
                    'TOV_rolling_3', 'FT_PCT_rolling_3', 'USG_PCT', 'TS','player_pos_C','player_pos_F', 'player_pos_G',
       'player_pos_PF', 'player_pos_SF']]

y_train = train_df['Fantasy_Point']

X_test = test_df[['PTS', 'MIN_x', 'TOV_x', 'AST_x', 'REB_x', 'FT_PCT', 
                  'PTS_rolling_3', 'AST_rolling_3', 'REB_rolling_3', 
                  'TOV_rolling_3', 'FT_PCT_rolling_3', 'USG_PCT', 'TS','player_pos_C','player_pos_F', 'player_pos_G',
       'player_pos_PF', 'player_pos_SF']]

y_test = test_df['Fantasy_Point']

y_train = train_df['Fantasy_Point']

# Extract player names for grouping
player_ids = train_df['Player Name']  # Used for GroupKFold grouping

from sklearn.model_selection import GroupKFold
from skopt import BayesSearchCV
from xgboost import XGBRegressor
from skopt.space import Real, Integer

# Define the model
model = XGBRegressor()

# Parameter grid for Bayesian optimization
param_grid = {
    'learning_rate': Real(0.01, 0.3, prior='log-uniform'),
    'max_depth': Integer(3, 10),
    'n_estimators': Integer(100, 1000),
    'subsample': Real(0.5, 1.0, prior='uniform'),
    'colsample_bytree': Real(0.5, 1.0, prior='uniform'),
    'reg_alpha': Real(0.01, 10, prior='log-uniform'),
    'reg_lambda': Real(0.01, 10, prior='log-uniform')
}

# Define GroupKFold for cross-validation
group_kfold = GroupKFold(n_splits=5)

# Setup Bayesian Search
bayes_search = BayesSearchCV(
    estimator=model,
    search_spaces=param_grid,
    scoring='neg_mean_squared_error',
    cv=group_kfold,
    n_iter=50,
    random_state=42,
    verbose=1
)

# Train the model with Bayesian Search
bayes_search.fit(
    X_train, 
    y_train, 
    groups=player_ids  # Provide group information here
)

# Print the best parameters and score
print("Best parameters:", bayes_search.best_params_)
print("Best score:", bayes_search.best_score_)

from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# Get the best parameters from Bayesian Search
best_params = bayes_search.best_params_

# Create a new model with the best parameters and train it
best_model = XGBRegressor(**best_params)
best_model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = best_model.predict(X_test)

# Calculate different metrics
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE
mae = mean_absolute_error(y_test, y_pred)  # MAE
r2 = r2_score(y_test, y_pred)  # RÂ²

# Filter non-zero indices for MAPE calculation
non_zero_indices = y_test != 0
y_test_filtered = y_test[non_zero_indices]
y_pred_filtered = y_pred[non_zero_indices]

# Calculate MAPE
mape = (abs((y_test_filtered - y_pred_filtered) / y_test_filtered).mean()) * 100
print(' Just Rolling')
print(f"MAPE: {mape}%")

# Print results
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"RÂ²: {r2}")
print(f"MAPE: {mape}%")

# Optionally, print feature importances
print("Feature Importances:")
print(best_model.feature_importances_)
```

# Ultimate Results after cross validation by using PCA APPLICATION

### The reason why we use pca is there was a multi dimensionality curse because rolling and normal points was highly correlated, which is why we applied pca method in order to improve performance of the model.

##### When you look at other outcomes at different combinations without pca, you will see that the model does not show any improvement despite k-fold CV and model tunning is implemented. After PCA implementation, the mape shows any improvement from %9 to %6. There is also an improvement on R2 how the model explains variance. 

```python
import pandas as pd
gbm_data=pd.read_csv('gbm_data.csv')
# get rolling averages for each player
gbm_data['PTS_rolling_3'] = gbm_data.groupby('Player_ID')['PTS'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data['REB_rolling_3'] = gbm_data.groupby('Player_ID')['REB_x'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data['AST_rolling_3'] = gbm_data.groupby('Player_ID')['AST_x'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data['USG_PCT_rolling_3'] = gbm_data.groupby('Player_ID')['USG_PCT'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data['FG_PCT_rolling_3'] = gbm_data.groupby('Player_ID')['FG_PCT'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data['FT_PCT_rolling_3'] = gbm_data.groupby('Player_ID')['FT_PCT'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data['TOV_rolling_3'] = gbm_data.groupby('Player_ID')['TOV_x'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data['TS_rolling_3'] = gbm_data.groupby('Player_ID')['TS'].transform(lambda x: x.rolling(3, min_periods=1).mean())
gbm_data
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler  
from sklearn.decomposition import PCA  

#Create datasets
train_df = pd.DataFrame()
test_df = pd.DataFrame()

# divide dataset according to player_id as 80:20 ratio
for player_id, group in gbm_data.groupby('Player_ID'):  
   
    split_index = int(len(group) * 0.8)
    train_part = group.iloc[:split_index]  # 0.80 for train
    test_part = group.iloc[split_index:]   # 0.20 for test
    
    # add divided player_id groups to main train and test dataframes
    train_df = pd.concat([train_df, train_part])
    test_df = pd.concat([test_df, test_part])
# adjust them as X and Y
X_train = train_df.drop(['Unnamed: 0', 'Player_ID', 'Player Name', 'Date'],axis=1)
y_train = train_df['Fantasy_Point']
X_test=test_df.drop(['Unnamed: 0', 'Player_ID', 'Player Name', 'Date'],axis=1)
y_test=test_df['Fantasy_Point']
scaler=StandardScaler()
# apply standartization for scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)  # It is learned on the training set and the transformation is applied
X_test_scaled = scaler.transform(X_test)  # Transformation with parameters learned from the training set

# PCA Hesaplama (EÄŸitim Seti)
pca = PCA(n_components=0.95)  # Determine the number of components that explain 95% of the variance
X_train_pca = pca.fit_transform(X_train_scaled)  # PCA is calculated on the training set

# Test Setine PCA Uygulama
X_test_pca = pca.transform(X_test_scaled)  # The PCA transformation learned from the training set is applied to the test set
# Test Setine PCA Uygulama
X_test_pca = pca.transform(X_test_scaled)  # The PCA transformation learned from the training set is applied to the test set

# Examine PCA result as DataFrame
pca_columns = [f"PCA_{i+1}" for i in range(X_train_pca.shape[1])]
X_train_pca_df = pd.DataFrame(X_train_pca, columns=pca_columns)
X_test_pca_df = pd.DataFrame(X_test_pca, columns=pca_columns)

# Print results
print("EÄŸitim Seti PCA:")
print(X_train_pca_df.head())
print("\nTest Seti PCA:")
print(X_test_pca_df.head())

#----------------------------------------------------------#
player_ids = train_df['Player Name']  # it will be used group kfold so that the dataset folds will keep together the same player names

from sklearn.model_selection import GroupKFold
from skopt import BayesSearchCV
from xgboost import XGBRegressor
from skopt.space import Real, Integer

# Create Model
model = XGBRegressor()

# definete parameters for xgboost and their intervals
param_grid = {
    'learning_rate': Real(0.01, 0.3, prior='log-uniform'),
    'max_depth': Integer(3, 10),
    'n_estimators': Integer(100, 1000),
    'subsample': Real(0.5, 1.0, prior='uniform'),
    'colsample_bytree': Real(0.5, 1.0, prior='uniform'),
    'reg_alpha': Real(0.01, 10, prior='log-uniform'),
    'reg_lambda': Real(0.01, 10, prior='log-uniform')
}

# GroupKFold definition
group_kfold = GroupKFold(n_splits=5)

# Create Bayesian Search
bayes_search = BayesSearchCV(
    estimator=model,
    search_spaces=param_grid,
    scoring='neg_mean_squared_error',
    cv=group_kfold,
    n_iter=50,
    random_state=42,
    verbose=1
)

# Train the model
bayes_search.fit(
    X_train_pca_df,
    y_train, 
    groups=player_ids  
)

# The best parameter and score
print("En iyi parametreler:", bayes_search.best_params_)
print("En iyi skor:", bayes_search.best_score_)
#--------------------------------------------------------------------
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

# get best params from bayesian search
best_params = bayes_search.best_params_

# apply them and train the model
best_model = XGBRegressor(**best_params)
best_model.fit(X_train_pca_df,y_train)

# prediction on test set
y_pred = best_model.predict(X_test_pca_df)

# calculate different metrics
rmse = mean_squared_error(y_test, y_pred, squared=False)  # RMSE
mae = mean_absolute_error(y_test, y_pred)  # MAE
r2 = r2_score(y_test, y_pred)  # RÂ²
non_zero_indices = y_test != 0
y_test_filtered = y_test[non_zero_indices]
y_pred_filtered = y_pred[non_zero_indices]

mape = (abs((y_test_filtered - y_pred_filtered) / y_test_filtered).mean()) * 100
print('without rolling')
print(f"MAPE: {mape}%")
# print results
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"RÂ²: {r2}")
print(f"MAPE: {mape}%")
# Optionally, you can also print the feature importances
print("Feature Importances:")
print(best_model.feature_importances_)
```

```python

cv_score = abs(bayes_search.best_score_)  
performance_diff = abs(cv_score - mape)

print(f"CV MAPE (Cross Validation): {cv_score:.4f}")
print(f"Test MAPE: {mape:.4f}")
print(f"CV-Test PerformansÄ± difference: {performance_diff:.4f}")

if performance_diff < 1.0:
    print("The model generalizes well")
elif performance_diff < 5.0:
    print("The model can be accepted in generalization.")
else:
    print('It could be a sign of overfitting.')
```
